```python
from PIL import Image
import numpy as np

image = Image.open('./smile.webp');
image.save("./smile.jpg")
image.thumbnail ((300, 300))
im = np.array(image)
print(image, image, image)
im_R = im.copy()
im_R[:, :, (1, 2)] = 0
# print(im_R)
im_G = im.copy()
im_G[:, :, (0, 2)] = 0
# print(im_G)
im_B = im.copy()
im_B[:, :, (0, 1)] = 0

im_RGB = np.concatenate((im_R, im_G, im_B), axis=1)

pil_img = Image.fromarray(im_RGB)
display(Image.fromarray(im))
display(pil_img)
```

```python
import random
arr = np.array([
    [[100,2,103],[0,2,4], [1,3,5]], 
    [[1,2,3],[0,2,4], [1,200,5]], 
    [[1,200,3],[0,2,4], [100,200,255]]
]);
print(arr.size, arr.shape)
for i in range(arr.shape[0]):
  for j in range(arr.shape[1]):
    arr[i][j][0] = random.randint(0,255) #R
    arr[i][j][1] = random.randint(0,255)  #G
    arr[i][j][2] = random.randint(0,255) #B
plt.imshow(arr)
```
