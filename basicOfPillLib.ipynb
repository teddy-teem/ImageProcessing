```python
#Import required library
from PIL import Image
#Open Image
im = Image.open("smile.jpg")
#Image rotate & show
rotatedImage = im.rotate(45); #don't change original image, return rotated image
display(rotatedImage)  # im.show() not work in colab. that why used display
```

```python
# resize
im.thumbnail ((300, 300)) # It changes the original image
display(im)
```

```python
def convert(src, to=".png"):
  if(src):
    image = Image.open(src)
    image.save("./new"+to)
    print("Saved to "+to+" format, refresh your dir.")
    return
  return print("src not defined")

convert("./smile.jpg", ".gif")
```
